```python
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("fetal_health 2.csv")
df["fetal_health"] = df["fetal_health"].astype("category")

# Choose key features
features = [
    "baseline value",
    "accelerations",
    "fetal_movement",
    "uterine_contractions",
    "prolongued_decelerations",
    "severe_decelerations"
]

# Create subplot grid
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15,10))
axes = axes.flatten()

for i, feature in enumerate(features):
    sns.boxplot(data=df, x="fetal_health", y=feature, ax=axes[i])
    axes[i].set_title(feature)

plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv("fetal_health 2.csv")
df["fetal_health"] = df["fetal_health"].astype("category")

# Choose key features
features = [
    "light_decelerations", "abnormal_short_term_variability",
    "percentage_of_time_with_abnormal_long_term_variability"

]

# Create subplot grid
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15,10))
axes = axes.flatten()

for i, feature in enumerate(features):
    sns.boxplot(data=df, x="fetal_health", y=feature, ax=axes[i])
    axes[i].set_title(feature)

plt.tight_layout()
plt.show()
```

useful features to consider (first intuition) - abnormal_short_term_variability, uterine_contractions, prolonged_decelerations, baseline_value, accelerations, percentage_of_time_with_abnormal_long_variability (but has lot of outliers in class 1)

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import (
    MaxAbsScaler,
    MinMaxScaler,
    Normalizer,
    PowerTransformer,
    QuantileTransformer,
    RobustScaler,
    StandardScaler,
    minmax_scale,
)
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression

```

```python
# checking which combination of features would be best for a classifier model

for features in [
    [
        "abnormal_short_term_variability",
        "uterine_contractions",
        "baseline value",
        "accelerations",
        "percentage_of_time_with_abnormal_long_term_variability", "prolongued_decelerations"
    ],
    [
        "percentage_of_time_with_abnormal_long_term_variability",
        "abnormal_short_term_variability",
        "accelerations"
    ],
    [
        "baseline value",
        "uterine_contractions",
        "light_decelerations",
        "severe_decelerations"
    ]
]:
  transformer = make_column_transformer((
      StandardScaler(), features),
                                        remainder="drop")
  pipeline = make_pipeline(transformer, KNeighborsClassifier(n_neighbors=10) )
  score = cross_val_score(pipeline, X=df.drop(columns=["fetal_health"]), y=df["fetal_health"], cv=10, scoring="f1_macro")
  print (features, score.mean())
```

```python
# playing around with more combinations of features in options 1 and 2

candidate_features = ['abnormal_short_term_variability', 'uterine_contractions',
                      'baseline value', 'percentage_of_time_with_abnormal_long_term_variability',
                      'accelerations', "prolongued_decelerations"]
```

```python
from itertools import combinations
for features in combinations(candidate_features, 5):
  feature_list = list(features)
  new_transformer = make_column_transformer((
      StandardScaler(), feature_list),
                                        remainder="drop")
  new_pipeline = make_pipeline(new_transformer, KNeighborsClassifier(n_neighbors=10) )

  scores = cross_val_score(new_pipeline, X=df.drop(columns=["fetal_health"]), y=df["fetal_health"],
                           cv=10, scoring="f1_macro")
  print (features, scores.mean())
```

```python
final_features = ['abnormal_short_term_variability', 'uterine_contractions',
                  'percentage_of_time_with_abnormal_long_term_variability', 'accelerations', 'prolongued_decelerations']
y = df["fetal_health"].astype(float).astype(int) - 1
X = df.drop(columns=["fetal_health"])
```

```python
# testing different models
y = df["fetal_health"].astype(int) - 1   # 1->0, 2->1, 3->2
X = df.drop(columns=["fetal_health"])

transformer = make_column_transformer((StandardScaler(), final_features), remainder="drop")

pipeline = make_pipeline(transformer, ('model'))

param_grid =[
    {
        "model": [KNeighborsClassifier()],
        "model__n_neighbors": [3, 5, 10, 20],
        "model__weights": ["uniform", "distance"]
    },

    {
        "model": [LogisticRegression(max_iter=5000)],
        "model__C": [0.1, 1, 10],
        "model__class_weight": [None, "balanced"]
    },

    {
        "model": [SVC()],
        "model__kernel": ["linear", "rbf"],
        "model__C": [0.1, 1, 10],
        "model__gamma": ["scale", "auto"]
    },

    {
        "model": [RandomForestClassifier(random_state=0)],
        "model__n_estimators": [200, 500],
        "model__max_depth": [None, 5, 10],
        "model__class_weight": [None, "balanced"]
    },

    {
        "model": [XGBClassifier(
            objective="multi:softprob",
            num_class=3,
            eval_metric="mlogloss",
            random_state=0
        )],
        "model__n_estimators": [200, 500],
        "model__max_depth": [3, 5],
        "model__learning_rate": [0.05, 0.1],
        "model__subsample": [0.8, 1.0],
        "model__colsample_bytree": [0.8, 1.0]
    }
]
grid_cv = GridSearchCV(pipeline,
                       param_grid=param_grid, cv=10, scoring="f1_macro", return_train_score=True)

grid_cv.fit(X, y)
grid_cv.best_params_
```

```python
pipeline = make_pipeline(
    transformer,
    XGBClassifier(
        colsample_bytree=1.0,
        learning_rate=0.05,
        max_depth=3,
        n_estimators=200,
        subsample=1.0,
        objective="multi:softprob",
        num_class=3,
        eval_metric="mlogloss",
        random_state=0
    )
)
(cross_val_score(pipeline, X, y, cv=10, scoring="f1_macro")).mean()
```

```python
# testing more xgboost hyperparameters

transformer = make_column_transformer((StandardScaler(), final_features), remainder="drop")

pipeline = make_pipeline(transformer, XGBClassifier())

param_grid =    {
    "xgbclassifier__n_estimators": [200, 500],
    "xgbclassifier__max_depth": [3, 5],
    "xgbclassifier__learning_rate": [0.05, 0.1],
    "xgbclassifier__subsample": [0.8, 1.0],
    "xgbclassifier__colsample_bytree": [0.8, 1.0],
    "xgbclassifier__min_child_weight": [1, 5],
    "xgbclassifier__reg_lambda": [1, 2],
}

grid_cv = GridSearchCV(pipeline,
                       param_grid=param_grid, cv=10, scoring="f1_macro", return_train_score=True)

grid_cv.fit(X, y)
grid_cv.best_params_
```

```python
transformer = make_column_transformer((StandardScaler(), final_features), remainder="drop")

pipeline = make_pipeline(
    transformer,
    XGBClassifier(
        colsample_bytree=0.8,
        learning_rate=0.05,
        max_depth=3,
        n_estimators=200,
        subsample=1.0,
        objective="multi:softprob",
        num_class=3,
        eval_metric="mlogloss",
        random_state=0,
        min_child_weight=1,
        reg_lambda=1
    )
)
(cross_val_score(pipeline, X, y, cv=10, scoring="f1_macro")).mean()
```

given this is the model with the highest f1 score so far, I will pivot from now to tuning for the highest recall for the pathological class

```python
from sklearn.metrics import make_scorer, recall_score

def recall_path(y_true, y_pred):
    return recall_score(y_true, y_pred, labels=[2], average="macro")

recall_path_scorer = make_scorer(recall_path)

cross_val_score(pipeline, X, y, cv=10, scoring=recall_path_scorer).mean()
```

```python
# tuning more XGBoost hyperparamaters
param_grid = {
    "xgbclassifier__learning_rate": [0.03, 0.05, 0.1],
    "xgbclassifier__colsample_bytree": [0.8, 1.0],
    "xgbclassifier__gamma": [0, 1, 5],
    "xgbclassifier__reg_alpha": [0, 0.1, 1.0],
}
grid = GridSearchCV(pipeline,
    param_grid=param_grid,
    cv=10,
    scoring=recall_path_scorer
)

grid.fit(X, y)
grid.best_params_
```

```python
pipeline.set_params(**grid.best_params_)
```

```python
(cross_val_score(pipeline, X, y, cv=10, scoring=recall_path_scorer)).mean()
```

```python
# best model so far, checking confusion matrix for this
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix, classification_report

pred = cross_val_predict(pipeline, X, y, cv=10)
print(confusion_matrix(y, pred))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay

class_names = ["Normal", "Suspect", "Pathological"]

disp = ConfusionMatrixDisplay.from_predictions(
    y,
    pred,
    display_labels=class_names,
    cmap="Blues",
    xticks_rotation=45
)

plt.title("Confusion Matrix (10-fold CV)")
plt.show()
```
